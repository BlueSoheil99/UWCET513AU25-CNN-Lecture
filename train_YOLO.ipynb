```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data="dataset-v2-1/data.yaml", epochs=100, imgsz=640)
```

When training is finished, you can find your model in 'runs/detect/train/weights/best.pt'
